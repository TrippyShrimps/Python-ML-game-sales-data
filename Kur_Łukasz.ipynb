Zbiór danych zawiera spis najlepiej sprzedających się gier wideo w liczbach ponad 100,000 sztuk na danej platformie. 
Pierwotnie pobrany zbiór z Keggle.com zawierał ponad 16tys wpisów, lecz w celu sproszczenia wybrałem top 25 najlepiej sprzedających się gier.

Zbiór jest podzielony na 11 kolumn:
    - Ranking na podstawie sprzedanych egzemplarzy,
    - Nazwa gry wideo,
    - Platforma wydania,
    - Rok wydania,
    - Gatunek,
    - Wydawca,
    - Sprzedaż w milionach w głównych regionach, reszcie świata oraz całości świata.

Biblioteką Pandas wyświetlono najważniejsze pozycje poszczególnych kolumn.

Dzięki NumPy obliczono średnią wartość globalnych sprzedaży przez 25 pozycji, jak i zakres, medianę oraz wariancję.

Następnie przesortowano tabelę przez rok wydania. Można sortować tabelę wedle preferecncji.

Ostatecznie wyświetla się tabelka porównująca sprzedaże Japonii do Europy.

```python
import pandas as pd 
df = pd.read_csv('vgsales.csv') #Dodanie pliku .csv
print(df.head()) #Wyświetlenie najważniejszych pozycji
```

```python
import numpy as np
average = np.mean(df['Global_Sales']) #Średnia wartość kolumny
zakres = np.ptp(df['Global_Sales']) #Różnica pomiędzy minimum a maximum kolumny
median = np.median(df['Global_Sales']) #Mediana 
wariancja = np.var(df['Global_Sales']) #Wariancja

print(average)
print(zakres)
print(median)
print(wariancja)
```

```python
df = df.sort_values('Year') #Sortowanie tabeli przez rok wydania
print(df.to_string())
```

```python
import matplotlib.pyplot as plt
plt.plot(df['JP_Sales'], df['EU_Sales']) #Sprzedaże Japonia i Europa
plt.xlabel('JP') #Podpis osi
plt.ylabel('EU')
plt.title('Sales comparison JP to EU') #Nagłówek
plt.show()
```
